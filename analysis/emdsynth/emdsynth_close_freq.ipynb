## Ejemplos de Uso

A continuación se muestran varios ejemplos de cómo usar la función con diferentes parámetros para evaluar el comportamiento de EMD con frecuencias muy cercanas.

```python
# Ejemplo 1: Frecuencias muy cercanas (f1=10, f2=12, razón = 0.83)
# Este caso es difícil para EMD según el límite de Rilling-Flandrin
t1, señal1 = generar_senal_frecuencia_cercana(
    f1=10.0,
    f2=12.0,
    amplitud1=1.0,
    amplitud2=1.0,
    duracion=2.0,
    frecuencia_muestreo=1000.0,
    funcion="cos"
)

fig = go.Figure()

# Añadir la señal
fig.add_trace(
    go.Scatter(
        x=t1,
        y=señal1,
        mode="lines",
        name="Señal completa",
        line=dict(color="blue", width=1.5),
    )
)

fig.update_layout(
    title="Señal con Frecuencias Muy Cercanas (f₁=10 Hz, f₂=12 Hz, razón=0.83)",
    xaxis_title="Tiempo (s)",
    yaxis_title="Amplitud",
    hovermode="x unified",
    width=1200,
    height=600,
    showlegend=True,
    template="plotly_white",
)

fig.show()

# Calcular y mostrar información sobre el batimiento
razon_frecuencias = min(10.0, 12.0) / max(10.0, 12.0)
frecuencia_batimiento = abs(12.0 - 10.0)
periodo_batimiento = 1.0 / frecuencia_batimiento if frecuencia_batimiento > 0 else np.inf

print(f"Razón de frecuencias: {razon_frecuencias:.3f}")
print(f"Frecuencia de batimiento: {frecuencia_batimiento:.3f} Hz")
print(f"Período de batimiento: {periodo_batimiento:.3f} s")
```

```python
# Ejemplo 2: Frecuencias aún más cercanas (f1=10, f2=10.5, razón = 0.95)
# Este caso es extremadamente difícil para EMD
t2, señal2 = generar_senal_frecuencia_cercana(
    f1=10.0,
    f2=10.5,
    amplitud1=1.0,
    amplitud2=1.0,
    duracion=4.0,
    frecuencia_muestreo=1000.0,
    funcion="cos"
)

fig = go.Figure()

# Añadir la señal
fig.add_trace(
    go.Scatter(
        x=t2,
        y=señal2,
        mode="lines",
        name="Señal completa",
        line=dict(color="red", width=1.5),
    )
)

fig.update_layout(
    title="Señal con Frecuencias Extremadamente Cercanas (f₁=10 Hz, f₂=10.5 Hz, razón=0.95)",
    xaxis_title="Tiempo (s)",
    yaxis_title="Amplitud",
    hovermode="x unified",
    width=1200,
    height=600,
    showlegend=True,
    template="plotly_white",
)

fig.show()

# Calcular información sobre el batimiento
razon_frecuencias = min(10.0, 10.5) / max(10.0, 10.5)
frecuencia_batimiento = abs(10.5 - 10.0)
periodo_batimiento = 1.0 / frecuencia_batimiento if frecuencia_batimiento > 0 else np.inf

print(f"Razón de frecuencias: {razon_frecuencias:.3f}")
print(f"Frecuencia de batimiento: {frecuencia_batimiento:.3f} Hz")
print(f"Período de batimiento: {periodo_batimiento:.3f} s")
```

```python
# Ejemplo 3: Comparación de diferentes razones de frecuencia
# Visualizar cómo cambia el patrón de batimiento según la cercanía de las frecuencias

fig = make_subplots(
    rows=3,
    cols=1,
    subplot_titles=[
        "Razón = 0.83 (f₁=10, f₂=12) - Difícil para EMD",
        "Razón = 0.91 (f₁=10, f₂=11) - Muy difícil para EMD",
        "Razón = 0.95 (f₁=10, f₂=10.5) - Extremadamente difícil",
    ],
    vertical_spacing=0.08,
    shared_xaxes=True,
)

configuraciones = [
    (10.0, 12.0, 2.0, "blue"),
    (10.0, 11.0, 3.0, "green"),
    (10.0, 10.5, 4.0, "red"),
]

for idx, (f1, f2, duracion, color) in enumerate(configuraciones):
    t, señal = generar_senal_frecuencia_cercana(
        f1=f1,
        f2=f2,
        amplitud1=1.0,
        amplitud2=1.0,
        duracion=duracion,
        frecuencia_muestreo=1000.0,
        funcion="cos"
    )
    
    razon = min(f1, f2) / max(f1, f2)
    
    fig.add_trace(
        go.Scatter(
            x=t,
            y=señal,
            mode="lines",
            name=f"Razón = {razon:.2f}",
            line=dict(color=color, width=1.5),
            showlegend=False,
        ),
        row=idx + 1,
        col=1,
    )

fig.update_xaxes(title_text="Tiempo (s)", row=3, col=1)
fig.update_yaxes(title_text="Amplitud", row=2, col=1)

fig.update_layout(
    title_text="Comparación de Señales con Diferentes Razones de Frecuencia",
    height=1000,
    width=1200,
    template="plotly_white",
    hovermode="x unified",
)

fig.show()
```

```python
# Ejemplo 4: Visualización de componentes separadas
# Mostrar las dos componentes individuales y su suma

t, señal = generar_senal_frecuencia_cercana(
    f1=10.0,
    f2=12.0,
    amplitud1=1.0,
    amplitud2=1.0,
    duracion=2.0,
    frecuencia_muestreo=1000.0,
    funcion="cos"
)

# Calcular componentes por separado para visualización
componente1 = np.cos(2 * np.pi * 10.0 * t)
componente2 = np.cos(2 * np.pi * 12.0 * t)

fig = make_subplots(
    rows=3,
    cols=1,
    subplot_titles=[
        "Componente 1: cos(2π·10·t)",
        "Componente 2: cos(2π·12·t)",
        "Señal Completa (suma de componentes)",
    ],
    vertical_spacing=0.08,
    shared_xaxes=True,
)

# Componente 1
fig.add_trace(
    go.Scatter(
        x=t,
        y=componente1,
        mode="lines",
        name="Componente 1",
        line=dict(color="blue", width=1.5),
        showlegend=False,
    ),
    row=1,
    col=1,
)

# Componente 2
fig.add_trace(
    go.Scatter(
        x=t,
        y=componente2,
        mode="lines",
        name="Componente 2",
        line=dict(color="red", width=1.5),
        showlegend=False,
    ),
    row=2,
    col=1,
)

# Señal completa
fig.add_trace(
    go.Scatter(
        x=t,
        y=señal,
        mode="lines",
        name="Señal completa",
        line=dict(color="green", width=1.5),
        showlegend=False,
    ),
    row=3,
    col=1,
)

fig.update_xaxes(title_text="Tiempo (s)", row=3, col=1)
fig.update_yaxes(title_text="Amplitud", row=2, col=1)

fig.update_layout(
    height=1000,
    width=1200,
    template="plotly_white",
    hovermode="x unified",
)

fig.show()
```

```python
# Ejemplo 5: Diferentes amplitudes para las componentes
# Evaluar el efecto de tener componentes con diferentes amplitudes

fig = make_subplots(
    rows=3,
    cols=1,
    subplot_titles=[
        "A₁=1.0, A₂=1.0 (amplitudes iguales)",
        "A₁=1.0, A₂=0.5 (segunda componente más débil)",
        "A₁=0.5, A₂=1.0 (primera componente más débil)",
    ],
    vertical_spacing=0.08,
    shared_xaxes=True,
)

configuraciones_amplitud = [
    (1.0, 1.0, "blue"),
    (1.0, 0.5, "green"),
    (0.5, 1.0, "red"),
]

for idx, (amp1, amp2, color) in enumerate(configuraciones_amplitud):
    t, señal = generar_senal_frecuencia_cercana(
        f1=10.0,
        f2=12.0,
        amplitud1=amp1,
        amplitud2=amp2,
        duracion=2.0,
        frecuencia_muestreo=1000.0,
        funcion="cos"
    )
    
    fig.add_trace(
        go.Scatter(
            x=t,
            y=señal,
            mode="lines",
            name=f"A₁={amp1}, A₂={amp2}",
            line=dict(color=color, width=1.5),
            showlegend=False,
        ),
        row=idx + 1,
        col=1,
    )

fig.update_xaxes(title_text="Tiempo (s)", row=3, col=1)
fig.update_yaxes(title_text="Amplitud", row=2, col=1)

fig.update_layout(
    title_text="Efecto de Diferentes Amplitudes en Señales con Frecuencias Cercanas",
    height=1000,
    width=1200,
    template="plotly_white",
    hovermode="x unified",
)

fig.show()
```

```python
# Ejemplo 6: Comparación entre cos y sin
# Mostrar la diferencia entre usar coseno y seno

t, señal_cos = generar_senal_frecuencia_cercana(
    f1=10.0,
    f2=12.0,
    amplitud1=1.0,
    amplitud2=1.0,
    duracion=2.0,
    frecuencia_muestreo=1000.0,
    funcion="cos"
)

t, señal_sin = generar_senal_frecuencia_cercana(
    f1=10.0,
    f2=12.0,
    amplitud1=1.0,
    amplitud2=1.0,
    duracion=2.0,
    frecuencia_muestreo=1000.0,
    funcion="sin"
)

fig = make_subplots(
    rows=2,
    cols=1,
    subplot_titles=[
        "Usando coseno: cos(2π·f₁·t) + cos(2π·f₂·t)",
        "Usando seno: sin(2π·f₁·t) + sin(2π·f₂·t)",
    ],
    vertical_spacing=0.08,
    shared_xaxes=True,
)

# Señal con coseno
fig.add_trace(
    go.Scatter(
        x=t,
        y=señal_cos,
        mode="lines",
        name="Coseno",
        line=dict(color="blue", width=1.5),
        showlegend=False,
    ),
    row=1,
    col=1,
)

# Señal con seno
fig.add_trace(
    go.Scatter(
        x=t,
        y=señal_sin,
        mode="lines",
        name="Seno",
        line=dict(color="red", width=1.5),
        showlegend=False,
    ),
    row=2,
    col=1,
)

fig.update_xaxes(title_text="Tiempo (s)", row=2, col=1)
fig.update_yaxes(title_text="Amplitud", row=1, col=1)

fig.update_layout(
    title_text="Comparación entre Coseno y Seno para Frecuencias Cercanas",
    height=800,
    width=1200,
    template="plotly_white",
    hovermode="x unified",
)

fig.show()
```

## Notas sobre el Límite de Rilling-Flandrin

El límite de Rilling-Flandrin establece que EMD puede separar correctamente dos componentes sinusoidales si la razón de frecuencias es mayor que aproximadamente 2. Para razones menores, especialmente cercanas a 1, se esperan problemas:

- **Razón > 2**: EMD debería separar correctamente las componentes
- **1 < Razón ≤ 2**: EMD puede tener dificultades
- **Razón ≈ 1**: EMD probablemente generará un solo IMF con batimiento

Los ejemplos anteriores muestran casos donde la razón es menor que 1, lo cual es especialmente desafiante para EMD. Estos casos son útiles para evaluar si variantes como EEMD o CEEMDAN pueden mejorar la separación.
