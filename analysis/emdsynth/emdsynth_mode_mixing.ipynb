```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from typing import Tuple, Optional
```

```python
from GraphEMD.data.emdsynth_utils import generate_mode_mixing_signal

# La función generate_mode_mixing_signal ahora se importa desde GraphEMD.data.emdsynth_utils
```

## Ejemplo de Uso

A continuación se muestra un ejemplo de cómo usar la función con diferentes parámetros:

```python
# Ejemplo 1: Señal con componente intermitente en el medio
t1, señal1 = generate_mode_mixing_signal(
    f_low=1.0,
    f_high=10.0,
    alpha=0.5,
    t1=3.0,
    t2=7.0,
    duracion=10.0,
    frecuencia_muestreo=1000.0
)

fig = go.Figure()

# Añadir región sombreada para el intervalo de alta frecuencia
fig.add_shape(
    type="rect",
    x0=3.0,
    x1=7.0,
    y0=min(señal1),
    y1=max(señal1),
    fillcolor="red",
    opacity=0.2,
    layer="below",
    line_width=0,
)

# Añadir la señal
fig.add_trace(
    go.Scatter(
        x=t1,
        y=señal1,
        mode="lines",
        name="Señal completa",
        line=dict(color="blue"),
    )
)

fig.update_layout(
    title="Señal Mode Mixing: Componente intermitente en el medio",
    xaxis_title="Tiempo (s)",
    yaxis_title="Amplitud",
    hovermode="x unified",
    width=1200,
    height=600,
    showlegend=True,
    template="plotly_white",
)

fig.show()
```

```python
# Ejemplo 2: Señal con componente intermitente al inicio
t2, señal2 = generate_mode_mixing_signal(
    f_low=0.5,
    f_high=15.0,
    alpha=0.8,
    t1=0.5,
    t2=2.5,
    duracion=10.0,
    frecuencia_muestreo=1000.0
)

fig = go.Figure()

# Añadir región sombreada para el intervalo de alta frecuencia
fig.add_shape(
    type="rect",
    x0=0.5,
    x1=2.5,
    y0=min(señal2),
    y1=max(señal2),
    fillcolor="red",
    opacity=0.2,
    layer="below",
    line_width=0,
)

# Añadir la señal
fig.add_trace(
    go.Scatter(
        x=t2,
        y=señal2,
        mode="lines",
        name="Señal completa",
        line=dict(color="blue"),
    )
)

fig.update_layout(
    title="Señal Mode Mixing: Componente intermitente al inicio",
    xaxis_title="Tiempo (s)",
    yaxis_title="Amplitud",
    hovermode="x unified",
    width=1200,
    height=600,
    showlegend=True,
    template="plotly_white",
)

fig.show()
```

```python
# Ejemplo 3: Comparación de diferentes amplitudes de la componente intermitente
# Ejemplo 3: Comparación de diferentes amplitudes de la componente intermitente
fig = make_subplots(
    rows=3,
    cols=1,
    subplot_titles=["Alpha = 0.3", "Alpha = 0.6", "Alpha = 1.0"],
    vertical_spacing=0.08,
)

alphas = [0.3, 0.6, 1.0]
for idx, alpha in enumerate(alphas):
    t, señal = generate_mode_mixing_signal(
        f_low=1.0,
        f_high=10.0,
        alpha=alpha,
        t1=3.0,
        t2=7.0,
        duracion=10.0,
        frecuencia_muestreo=1000.0
    )
    
    # Añadir región sombreada
    fig.add_shape(
        type="rect",
        x0=3.0,
        x1=7.0,
        y0=min(señal),
        y1=max(señal),
        fillcolor="red",
        opacity=0.2,
        layer="below",
        line_width=0,
        row=idx + 1,
        col=1,
    )
    
    # Añadir la señal
    fig.add_trace(
        go.Scatter(
            x=t,
            y=señal,
            mode="lines",
            name=f"Alpha = {alpha}",
            line=dict(color="blue"),
            showlegend=False,
        ),
        row=idx + 1,
        col=1,
    )

fig.update_xaxes(title_text="Tiempo (s)", row=3, col=1)
fig.update_yaxes(title_text="Amplitud", row=2, col=1)

fig.update_layout(
    title_text="Comparación de diferentes amplitudes de componente intermitente",
    height=1000,
    width=1200,
    template="plotly_white",
    hovermode="x unified",
)

fig.show()
```

```python
# Ejemplo 4: Visualización de componentes separadas
t, señal = generate_mode_mixing_signal(
    f_low=1.0,
    f_high=10.0,
    alpha=0.5,
    t1=3.0,
    t2=7.0,
    duracion=10.0,
    frecuencia_muestreo=1000.0
)

# Calcular componentes por separado para visualización
componente_baja = np.sin(2 * np.pi * 1.0 * t)
indicador = (t >= 3.0) & (t <= 7.0)
componente_alta = 0.5 * np.sin(2 * np.pi * 10.0 * t) * indicador

fig = make_subplots(
    rows=3,
    cols=1,
    subplot_titles=[
        "Componente de Baja Frecuencia (continua)",
        "Componente de Alta Frecuencia (intermitente)",
        "Señal Completa (suma de componentes)",
    ],
    vertical_spacing=0.08,
    shared_xaxes=True,
)

# Componente baja frecuencia
fig.add_trace(
    go.Scatter(
        x=t,
        y=componente_baja,
        mode="lines",
        name="Componente baja frecuencia",
        line=dict(color="blue"),
        showlegend=False,
    ),
    row=1,
    col=1,
)

# Componente alta frecuencia con región sombreada
fig.add_shape(
    type="rect",
    x0=3.0,
    x1=7.0,
    y0=min(componente_alta),
    y1=max(componente_alta),
    fillcolor="red",
    opacity=0.2,
    layer="below",
    line_width=0,
    row=2,
    col=1,
)

fig.add_trace(
    go.Scatter(
        x=t,
        y=componente_alta,
        mode="lines",
        name="Componente alta frecuencia",
        line=dict(color="red"),
        showlegend=False,
    ),
    row=2,
    col=1,
)

# Señal completa con región sombreada
fig.add_shape(
    type="rect",
    x0=3.0,
    x1=7.0,
    y0=min(señal),
    y1=max(señal),
    fillcolor="red",
    opacity=0.2,
    layer="below",
    line_width=0,
    row=3,
    col=1,
)

fig.add_trace(
    go.Scatter(
        x=t,
        y=señal,
        mode="lines",
        name="Señal completa",
        line=dict(color="green"),
        showlegend=False,
    ),
    row=3,
    col=1,
)

fig.update_xaxes(title_text="Tiempo (s)", row=3, col=1)
fig.update_yaxes(title_text="Amplitud", row=2, col=1)

fig.update_layout(
    height=1000,
    width=1200,
    template="plotly_white",
    hovermode="x unified",
)

fig.show()
```
