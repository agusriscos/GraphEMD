```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

from GraphEMD.data.emdsynth_utils import generar_senal_chirp
```

```python
# La función generar_senal_chirp ahora se importa desde GraphEMD.data.emdsynth_utils
```

## Ejemplos de Uso

A continuación se muestran varios ejemplos de cómo usar la función con diferentes parámetros para evaluar el comportamiento de EMD con señales chirp.

```python
# Chirp ascendente: frecuencia inicial 1 Hz, aumenta a 6 Hz en 10 segundos
t1, señal1 = generar_senal_chirp(
    f0=1.0,
    k=0.5,  # La frecuencia aumenta 0.5 Hz por segundo
    duracion=10.0,
    frecuencia_muestreo=1000.0
)

# Calcular frecuencia instantánea para visualización
frecuencia_instantanea1 = 1.0 + 0.5 * t1

# Crear visualización
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=('Señal Chirp Ascendente', 'Frecuencia Instantánea'),
    vertical_spacing=0.12
)

# Señal temporal
fig.add_trace(
    go.Scatter(
        x=t1,
        y=señal1,
        mode='lines',
        name='Señal chirp',
        line=dict(color='blue', width=1.5)
    ),
    row=1, col=1
)

# Frecuencia instantánea
fig.add_trace(
    go.Scatter(
        x=t1,
        y=frecuencia_instantanea1,
        mode='lines',
        name='Frecuencia instantánea',
        line=dict(color='red', width=2)
    ),
    row=2, col=1
)

fig.update_xaxes(title_text="Tiempo (s)", row=1, col=1)
fig.update_xaxes(title_text="Tiempo (s)", row=2, col=1)
fig.update_yaxes(title_text="Amplitud", row=1, col=1)
fig.update_yaxes(title_text="Frecuencia (Hz)", row=2, col=1)

fig.update_layout(
    height=700,
    title_text="Ejemplo 1: Chirp Ascendente (f₀=1 Hz, k=0.5 Hz/s)",
    showlegend=True
)

fig.show()
```

## Ejemplos de Uso

## Ejemplos de Uso

A continuación se muestran varios ejemplos de cómo usar la función con diferentes parámetros para evaluar el comportamiento de EMD con señales chirp.

```python
# Generar múltiples señales con diferentes tasas de barrido
tasas_barrido = [0.1, 0.5, 1.0, 2.0]
colores = ['blue', 'green', 'orange', 'red']
nombres = [f'k = {k} Hz/s' for k in tasas_barrido]

fig = go.Figure()

for k, color, nombre in zip(tasas_barrido, colores, nombres):
    t, señal = generar_senal_chirp(
        f0=1.0,
        k=k,
        duracion=10.0,
        frecuencia_muestreo=1000.0
    )
    
    # Mostrar solo una porción para mejor visualización
    indices = slice(0, len(t), 10)  # Submuestrear para visualización
    
    fig.add_trace(
        go.Scatter(
            x=t[indices],
            y=señal[indices],
            mode='lines',
            name=nombre,
            line=dict(color=color, width=1.5)
        )
    )

fig.update_layout(
    title='Comparación de Señales Chirp con Diferentes Tasas de Barrido',
    xaxis_title='Tiempo (s)',
    yaxis_title='Amplitud',
    height=600,
    hovermode='x unified'
)

fig.show()
```

## Ejemplos de Uso

A continuación se muestran varios ejemplos de cómo usar la función con diferentes parámetros para evaluar el comportamiento de EMD con señales chirp.

```python
# Chirp ascendente: de 1 Hz a 6 Hz
t_asc, señal_asc = generar_senal_chirp(
    f0=1.0,
    k=0.5,
    duracion=10.0,
    frecuencia_muestreo=1000.0
)

# Chirp descendente: de 6 Hz a 1 Hz
t_desc, señal_desc = generar_senal_chirp(
    f0=6.0,
    k=-0.5,
    duracion=10.0,
    frecuencia_muestreo=1000.0
)

# Calcular frecuencias instantáneas
f_asc = 1.0 + 0.5 * t_asc
f_desc = 6.0 - 0.5 * t_desc

# Crear visualización
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=('Señales Chirp', 'Frecuencias Instantáneas'),
    vertical_spacing=0.12
)

# Señales
fig.add_trace(
    go.Scatter(
        x=t_asc,
        y=señal_asc,
        mode='lines',
        name='Chirp ascendente (1→6 Hz)',
        line=dict(color='blue', width=1.5)
    ),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(
        x=t_desc,
        y=señal_desc,
        mode='lines',
        name='Chirp descendente (6→1 Hz)',
        line=dict(color='red', width=1.5)
    ),
    row=1, col=1
)

# Frecuencias instantáneas
fig.add_trace(
    go.Scatter(
        x=t_asc,
        y=f_asc,
        mode='lines',
        name='Frecuencia ascendente',
        line=dict(color='blue', width=2, dash='dash')
    ),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(
        x=t_desc,
        y=f_desc,
        mode='lines',
        name='Frecuencia descendente',
        line=dict(color='red', width=2, dash='dash')
    ),
    row=2, col=1
)

fig.update_xaxes(title_text="Tiempo (s)", row=1, col=1)
fig.update_xaxes(title_text="Tiempo (s)", row=2, col=1)
fig.update_yaxes(title_text="Amplitud", row=1, col=1)
fig.update_yaxes(title_text="Frecuencia (Hz)", row=2, col=1)

fig.update_layout(
    height=700,
    title_text="Ejemplo 5: Comparación Chirp Ascendente vs Descendente",
    showlegend=True
)

fig.show()
```

```python
# Generar señales con diferentes amplitudes y fases
configuraciones = [
    {'amplitud': 1.0, 'fase': 0.0, 'nombre': 'A=1.0, φ=0', 'color': 'blue'},
    {'amplitud': 2.0, 'fase': 0.0, 'nombre': 'A=2.0, φ=0', 'color': 'green'},
    {'amplitud': 1.0, 'fase': np.pi/2, 'nombre': 'A=1.0, φ=π/2', 'color': 'orange'},
    {'amplitud': 1.0, 'fase': np.pi, 'nombre': 'A=1.0, φ=π', 'color': 'red'},
]

fig = go.Figure()

for config in configuraciones:
    t, señal = generar_senal_chirp(
        f0=2.0,
        k=0.3,
        amplitud=config['amplitud'],
        fase=config['fase'],
        duracion=5.0,
        frecuencia_muestreo=1000.0
    )
    
    # Submuestrear para mejor visualización
    indices = slice(0, len(t), 5)
    
    fig.add_trace(
        go.Scatter(
            x=t[indices],
            y=señal[indices],
            mode='lines',
            name=config['nombre'],
            line=dict(color=config['color'], width=1.5)
        )
    )

fig.update_layout(
    title='Ejemplo 6: Efecto de Amplitud y Fase en Señales Chirp',
    xaxis_title='Tiempo (s)',
    yaxis_title='Amplitud',
    height=600,
    hovermode='x unified'
)

fig.show()
```

## Ejemplos de Uso

A continuación se muestran varios ejemplos de cómo usar la función con diferentes parámetros para evaluar el comportamiento de EMD con señales chirp.

```python
# Generar señal con k=0 (frecuencia constante)
t_const, señal_const = generar_senal_chirp(
    f0=5.0,
    k=0.0,  # Sin cambio de frecuencia
    duracion=2.0,
    frecuencia_muestreo=1000.0
)

# Comparar con una sinusoide estándar para verificar
t_std = np.linspace(0, 2.0, 2000)
señal_std = np.sin(2 * np.pi * 5.0 * t_std)

# Crear visualización
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=('Señal con k=0 (frecuencia constante)', 'Comparación con sinusoide estándar'),
    vertical_spacing=0.15
)

# Señal chirp con k=0
fig.add_trace(
    go.Scatter(
        x=t_const,
        y=señal_const,
        mode='lines',
        name='Chirp con k=0',
        line=dict(color='blue', width=1.5)
    ),
    row=1, col=1
)

# Comparación
fig.add_trace(
    go.Scatter(
        x=t_const,
        y=señal_const,
        mode='lines',
        name='Chirp con k=0',
        line=dict(color='blue', width=1.5)
    ),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(
        x=t_std,
        y=señal_std,
        mode='lines',
        name='sin(2π·5·t) estándar',
        line=dict(color='red', width=1.5, dash='dash')
    ),
    row=2, col=1
)

fig.update_xaxes(title_text="Tiempo (s)", row=1, col=1)
fig.update_xaxes(title_text="Tiempo (s)", row=2, col=1)
fig.update_yaxes(title_text="Amplitud", row=1, col=1)
fig.update_yaxes(title_text="Amplitud", row=2, col=1)

fig.update_layout(
    height=700,
    title_text="Ejemplo 8: Caso Límite - Frecuencia Constante (k=0, f₀=5 Hz)",
    showlegend=True
)

fig.show()

# Verificar que son iguales (deberían ser idénticas)
diferencia = np.max(np.abs(señal_const[:2000] - señal_std))
print(f"Diferencia máxima entre señales: {diferencia:.2e}")
print("(Valores muy pequeños indican que son prácticamente idénticas)")
```
